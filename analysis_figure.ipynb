```python
import numpy as np
import os
import matplotlib.pyplot as plt
import cv2
import pandas as pd
```

# Different  backbone

```python
expcodes = ['08','09','10','07','11','12','13']
expbones = ['resnet18','resnet50','resnext50','efficientb0','efficientb1','efficientb2','efficientb3']
epoch = 50
list_color = ['k','r','b','y','g','brown','m']
file_path = 'exp_result/csv'

val_loss = []
train_loss = []
val_iou = []
train_iou = []

for exp in expcodes:
    csv_file =  f'{file_path}/exp_{exp}_history.csv'
    df = pd.read_csv(csv_file)
    val_loss.append(df['val_loss'].to_list())
    train_loss.append(df['train_loss'].to_list()) 
    val_iou.append(df['val_iou'].to_list())
    train_iou.append(df['train_iou'].to_list()) 
    
# plot val_loss
filename = 'exp_result/figure/analysis/val_loss_backbone.jpeg'
plt.figure(figsize=(15,10))
x_ticks = np.linspace(1,epoch,epoch)
for i in range(len(expcodes)):
    plt.plot(x_ticks,val_loss[i], color=list_color[i], label= expbones[i])
plt.title('val_loss(Unet) Model backbone', fontsize=20)
plt.xlabel('Epoch',fontsize=15)
plt.ylabel('loss',fontsize=15)

plt.xlim((1,epoch))
plt.ylim((0.8,1.0))
plt.xticks(x_ticks)
plt.grid()
plt.legend(prop={'size': 15})
plt.savefig(filename)

# plot train_loss
filename = 'exp_result/figure/analysis/train_loss_backbone.jpeg'
plt.figure(figsize=(15,10))
x_ticks = np.linspace(1,epoch,epoch)
for i in range(len(expcodes)):
    plt.plot(x_ticks,train_loss[i], color=list_color[i], label= expbones[i])
plt.title('train_loss(Unet) Model backbone', fontsize=20)
plt.xlabel('Epoch',fontsize=15)
plt.ylabel('loss',fontsize=15)

plt.xlim((1,epoch))
plt.ylim((0.7,0.9))
plt.xticks(x_ticks)
plt.grid()
plt.legend(prop={'size': 15})
plt.savefig(filename)

# plot val_iou
filename = 'exp_result/figure/analysis/val_iou_backbone.jpeg'
plt.figure(figsize=(15,10))
x_ticks = np.linspace(1,epoch,epoch)
for i in range(len(expcodes)):
    plt.plot(x_ticks,val_iou[i], color=list_color[i], label= expbones[i])
plt.title('val_iou(Unet) Model backbone', fontsize=20)
plt.xlabel('Epoch',fontsize=15)
plt.ylabel('loss',fontsize=15)

plt.xlim((1,epoch))
plt.ylim((0.2,0.5))
plt.xticks(x_ticks)
plt.grid()
plt.legend(prop={'size': 15})
plt.savefig(filename)

# plot train_iou
filename = 'exp_result/figure/analysis/train_iou_backbone.jpeg'
plt.figure(figsize=(15,10))
x_ticks = np.linspace(1,epoch,epoch)
for i in range(len(expcodes)):
    plt.plot(x_ticks,train_iou[i], color=list_color[i], label= expbones[i])
plt.title('train_iou(Unet) Model backbone', fontsize=20)
plt.xlabel('Epoch',fontsize=15)
plt.ylabel('loss',fontsize=15)

plt.xlim((1,epoch))
plt.ylim((0.4,0.55))
plt.xticks(x_ticks)
plt.grid()
plt.legend(prop={'size': 15})
plt.savefig(filename)
```
